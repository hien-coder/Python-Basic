# 1. Tông Quát

- except: chạy khi có lỗi
- else: khi không có lỗi xảy ra
- finally: luôn chạy

```python
while True:
    try:
        a = int(input("Nhập a: "))

        if a == 0:
            break

        print(f'{a}')
        
    except Exception as Ex:
        print("Error: ", Ex)

    else:
        print("Else chạy ...!")

    finally:
        print("finally chạy ...!")
```

# 2. Built-in exceptions
https://pythonbasics.org/try-except/#Built-in-exceptions

```python
try:
    a = int(input("Nhập a: "))
    b = int(input("Nhập a: "))

    print(f'{a} / {b} = {a/b}')

# chỉ bắt lỗi ValueError
except ValueError as Ve:
    print("Error: ", Ve)
```

# 3. Raise Exception

```python
def EnterA():
    try:
        a = int(input("Nhập a: "))
        print(a)

        if a == 0:
            raise ValueError
    except ValueError:
        raise ValueError("Lỗi giá trị nhập")
```

```python
try:
    EnterA()
except Exception as Ex:
    print(Ex)
```
